Compare Spacy and FrenchLefffLemmatizer

```python
# read file ..data/clean/Emea/Emea_test_500.tok.true.clean.fr

import spacy
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from french_lefff_lemmatizer.french_lefff_lemmatizer import FrenchLefffLemmatizer
lemmatizer_fr = FrenchLefffLemmatizer()

nlp = spacy.load('fr_core_news_md')

def get_char_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return 'a'
    elif treebank_tag.startswith('V'):
        return 'v'
    elif treebank_tag.startswith('N'):
        return 'n'
    elif treebank_tag.startswith('R'):
        return 'r'
    else:
        return None
    
with open("../data/clean/Emea/Emea_test_500.tok.true.clean.fr") as f:
    lines = f.readlines()
    for line in lines:
        line = line.lower().strip()
        words = word_tokenize(line)
        tagged_words = pos_tag(words)


        lemmatized_words = []
        for word, pos in tagged_words:
            pos = get_char_pos(pos)
            if pos:
                lemma = lemmatizer_fr.lemmatize(word, pos)
            else:
                lemma = word
            lemmatized_words.append(lemma)

        doc = nlp(line)

        for token, word in zip(doc, lemmatized_words):
            if token.lemma_  != word:
                print("origin: ", token.text ,"spacy: ", token.lemma_ , "nltk: ", word)

```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet


lemmatizer_en = WordNetLemmatizer()
nltk.download('wordnet')

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None
    
with open("../data/clean/Emea/Emea_test_500.tok.true.clean.en") as f:
    lines = f.readlines()
    for line in lines:
        line = line.lower().strip()
        words = word_tokenize(line)
        tagged_words = pos_tag(words)


        lemmatized_words = []
        for word, pos in tagged_words:
            pos = get_wordnet_pos(pos)
            if pos:
                lemma = lemmatizer_en.lemmatize(word, pos)
            else:
                lemma = word
            lemmatized_words.append(lemma)

        doc = nlp(line)

        for token, word in zip(doc, lemmatized_words):
            if token.lemma_  != word:
                print("origin: ", token.text ,"spacy: ", token.lemma_ , "nltk: ", word)
```
