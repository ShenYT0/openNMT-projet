```python
!onmt_translate -model ../data/model/run1_step_10000.pt -src ../data/clean/Europarl/Europarl_test_500.tok.true.clean.en -output ../data/result/pred1_Europarl_100000.txt -gpu 0 -verbose
```

```python
!onmt_translate -model ../data/model/run1_step_10000.pt -src ../data/clean/Emea/Emea_test_500.tok.true.clean.en -output ../data/result/pred1_Emea_100000.txt -gpu 0 -verbose
```

```python
import sacrebleu


with open("../data/result/pred1_Europarl_100000.txt", "r", encoding="utf-8") as pred_file:
    translations = pred_file.readlines()
with open("../data/clean/Europarl/Europarl_test_500.tok.true.clean.fr", "r", encoding="utf-8") as ref_file:
    references = ref_file.readlines()
translations = [str(trans) for trans in translations]
bleu_score = sacrebleu.corpus_bleu(references, translations)
print("BLEU Score run 1 Europarl:", bleu_score)


with open("../data/result/pred1_Emea_100000.txt", "r", encoding="utf-8") as pred_file:
    translations = pred_file.readlines()
with open("../data/clean/Emea/Emea_test_500.tok.true.clean.fr", "r", encoding="utf-8") as ref_file:
    references = ref_file.readlines()
translations = [str(trans) for trans in translations]
bleu_score = sacrebleu.corpus_bleu(references, translations)
print("BLEU Score run 1 Emea:", bleu_score)
```
